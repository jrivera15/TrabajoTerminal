# 1.1 Preprocesamiento de datos

En esta libreta se realiza la limpieza y preprocesamiento de la tabla 'IndicadorObtenerBiblia' para su posterior uso.

```python
import numpy as np
import pandas as pd
import os
```

```python
# display all columns
pd.set_option('display.max_columns', None)
```

```python
path = "../SGAData/"
df_raw = pd.read_csv(os.path.join(path, 'IndicadorObtenerBiblia.csv'), dtype={"CDPOSTAL": str})
df_raw
```

### LIMPIEZA

```python
df_clean = df_raw.copy()
```

#### COLUMNAS VACÍAS O IRRELEVANTES

```python
# Eliminar columnas completamente vacías
df_clean.dropna(axis=1, how='all', inplace=True)

# Eliminar columnas completamente con valor 0
df_clean.drop(columns=['EXPEDIENTE'], inplace=True)

# Eliminar columnas completamente con el valor "NO DISPONIBLE"
df_clean.drop(columns=['PAISTUTOR'], inplace=True)
df_clean.drop(columns=['ESTADOTUTOR2'], inplace=True)
df_clean.drop(columns=['CIUDADTUTOR'], inplace=True)
```

#### MUNALUMN

```python
# Homologación a 'BAHIA DE KINO'
df_clean['MUNALUMN'] = df_clean['MUNALUMN'].str.replace('BAHIADEKINO', 'BAHIA DE KINO')

# Homologación a 'HERMOSILLO'
df_clean['MUNALUMN'] = df_clean['MUNALUMN'].str.replace('HERMOISLLO', 'HERMOSILLO')
df_clean['MUNALUMN'] = df_clean['MUNALUMN'].str.replace('HERMSILLO', 'HERMOSILLO')
df_clean['MUNALUMN'] = df_clean['MUNALUMN'].str.replace('HERMOSIILLO', 'HERMOSILLO')

# Homologación a 'MIGUEL ALEMAN'
df_clean['MUNALUMN'] = df_clean['MUNALUMN'].str.replace('COSTA DE HERMOSILLO', 'MIGUEL ALEMAN')
df_clean['MUNALUMN'] = df_clean['MUNALUMN'].str.replace('POBLADO MIGUEL ALEMAN', 'MIGUEL ALEMAN')
df_clean['CIUDADPREPA'] = df_clean['CIUDADPREPA'].str.replace('POBLADO MIGUEL ALEMAN', 'MIGUEL ALEMAN')

# Homologación a 'CARBO'
df_clean['MUNALUMN'] = df_clean['MUNALUMN'].str.replace('CARBO SONORA', 'CARBO')

# Homologación a 'NAVOJOA'
df_clean['MUNALUMN'] = df_clean['MUNALUMN'].str.replace('NAVOJOA SONORA', 'NAVOJOA')

# Homologación 'OBREGON'
df_clean['MUNALUMN'] = df_clean['MUNALUMN'].str.replace('OBRGON', 'OBREGON')

# Homologación a 'CABORCA'
df_clean['MUNALUMN'] = df_clean['MUNALUMN'].str.replace('H CABORCA', 'CABORCA')
df_clean['MUNALUMN'] = df_clean['MUNALUMN'].str.replace('83600', 'CABORCA')

# Homologación a NaN para valores no válidos
mun_no_validos = ["NINGUNA", "NINGUNO", "UNICO DOMICILIO", "5", "NO", "6624", "SONORA", "NADA",
                  "NO HAY", "VIVO CON MIS PADRES", "VIVO CON MI MADRÉ", "ABUELOS"]
df_clean["MUNALUMN"] = df_clean["MUNALUMN"].replace(mun_no_validos, pd.NA)
```

#### CODIGO POSTAL

```python
# Añadir ceros a la izquierda en los CP del Distrito Federal
df_clean['CDPOSTAL'] = df_clean.apply(lambda row: row['CDPOSTAL'].zfill(5) if pd.isna(row['CDPOSTAL']) == False and row['ESTADOALUMN'] == "DISTRITO FEDERAL" else row['CDPOSTAL'], axis=1)
```

```python
# Imputar valores faltantes con '0'
df_clean['CDPOSTAL'] = df_clean['CDPOSTAL'].fillna('0').astype(str)

# Reemplazar NaN por 0 y convertir a entero
#df_clean['CDPOSTAL'] = df_clean['CDPOSTAL'].fillna(0).astype(int)

# Convertir a string para validar que solo tenga 5 dígitos, sino asignar "0", y convertir a entero
#df_clean['CDPOSTAL'] = df_clean['CDPOSTAL'].astype(str).where(lambda x: x.str.match(r'^\d{5}$'), "0").astype(int)
```

```python
# Crear una columna donde el CP sea de tipo numérico para posterior uso
df_clean['CDPOSTALNUM'] = df_clean['CDPOSTAL'].astype("Int64")
```

```python
#CPs en HERMOSILLO: 83000 - 83357 (zonas urbanas y rurales)
#CPs en SONORA: 83000 - 85994 
#CPs en MEXICO: 20000 - 99998
#CP en CDMX: 01000 - 16910
```

#### DUPLICADOS

Se eliminan registros duplicados, es decir, nos quedamos con un registro por estudiante. En caso de que un estudiante tenga múltiples registros debido a varias presentaciones del EGEL, nos quedaremos con el registro que muestre 'EGELDICTAMEN' = 'SATISFACTORIO'.

```python
df_sorted = df_clean.sort_values(by=['IDEstudiante', 'EGELPUNTAJEFINAL'], ascending=[True, False])
df_sorted
```

```python
df_nodups = df_sorted.drop_duplicates(subset='IDEstudiante', keep='first')
df_nodups
```

```python
# Guardar en archivo CSV
path = "../data/processed/"
df_nodups.to_csv(os.path.join(path,'IndicadorObtenerBiblia_clean.csv'), index=False)
```

### CATALOGOS

Creación de algunos catálogos para una mejor comprensión de los datos

```python
catalogo_ur = df_nodups.drop_duplicates(subset=['UR', 'NOMBREUNIDADREGIONAL'])[['UR', 'NOMBREUNIDADREGIONAL']]
catalogo_campus = df_nodups.drop_duplicates(subset=['CAMPUS'])[['IDCAMPUS','CAMPUS']]
catalogo_div = df_nodups.drop_duplicates(subset=['DIV', 'NOMBREDIVISION'])[['DIV', 'NOMBREDIVISION']]
catalogo_carrera = df_nodups.drop_duplicates(subset=['CARRERA', 'NOMBRECARRERA'])[['CARRERA', 'NOMBRECARRERA']]

# Guardar catálogos en CSV
path = "../data/processed/"
catalogo_ur.to_csv(os.path.join(path,'cat_unidadregional.csv'), index=False)
catalogo_campus.to_csv(os.path.join(path,'cat_campus.csv'), index=False)
catalogo_div.to_csv(os.path.join(path,'cat_division.csv'), index=False)
catalogo_carrera.to_csv(os.path.join(path,'cat_carrera.csv'), index=False)
```
