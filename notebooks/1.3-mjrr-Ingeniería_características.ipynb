## Tabla: IndicadorObtenerBiblia

```python
path = "../data/processed/"
df_indicador = pd.read_csv(os.path.join(path, 'IndicadorObtenerBiblia_geo.csv'),dtype={"CDPOSTAL": str})
df_indicador
```

### LUGAR DE ORIGEN

Determinar el lugar de origen del alumno (ya sea local o foráneo) ya que en algunas ocasiones este dato no se encuentra disponible, pero se puede deducir de otras columnas. Orden de prioridad:
1. CIUDADPREPA
2. ESTADOPREPA
3. MUNALUMN
4. ESTADOALUMN
5. PAISPREPA
6. PAIS

```python
df_indicador['LUGARORIGEN'] = df_indicador.apply(
    lambda row: 
        #         
        row['CIUDADPREPA'] if (pd.isna(row['CIUDADPREPA']) == False and row['CIUDADPREPA'] != 'NO DISPONIBLE') else
        row['ESTADOPREPA'] if (pd.isna(row['ESTADOPREPA']) == False and row['ESTADOPREPA'] != 'NO DISPONIBLE') else

        row['MUNALUMN'] if (pd.isna(row['MUNALUMN']) == False and row['MUNALUMN'] != 'NO DISPONIBLE') else
        row['ESTADOALUMN'] if (pd.isna(row['ESTADOALUMN']) == False and row['ESTADOALUMN'] != 'NO DISPONIBLE') else
            
        row['PAISPREPA'] if (pd.isna(row['PAISPREPA']) == False and row['PAISPREPA'] != 'NO DISPONIBLE') else
        row['PAIS'] if (pd.isna(row['PAIS']) == False and row['PAIS'] != 'NO DISPONIBLE') else        
        None,
    axis=1
)
```

```python
# checar casos: 
# MUNALUMN == SAN PEDRO EL SAUCITO y CIUDADPREPA == HERMOSILLO
# MUNALUMN == ALAMOS y CIUDADPREPA == hermosillo
df_indicador[(df_indicador['MUNALUMN'] != df_indicador['CIUDADPREPA']) & (df_indicador['CIUDADPREPA']!= 'NO DISPONIBLE')]
```

### FORANEO

Determinar si el alumno es foráneo según las siguientes condiciones:

```python
df_indicador['FORANEO'] = df_indicador.apply(   
    lambda row: 1 if row['LUGARORIGEN'] != 'HERMOSILLO' else 0,
    axis=1
)
```

### TITULADO

```python
df_indicador['TITULADO'] = df_indicador.apply(   
    lambda row: 1 if pd.isna(row['FECHATITULACION']) == False else 0,
    axis=1
)
```

### NÚMERO DE SEMESTRES CURSADOS

```python
path = "../SGAData/"
df_semestres_cursados = pd.read_csv(os.path.join(path, 'EstudianteSemestresCursados.csv'))

# Anexar columna de semestres cursados
df_indicador = df_indicador.merge(df_semestres_cursados[['IDEstudiante','SemestresCursados']], how='left', on='IDEstudiante')
```

### DESERCIÓN

```python
def identificar_desercion(cicloinicio, ultciclo, cicloegreso): 
    try:    
        ciclo_segundo_semestre = None  
      
        # Se calcula el ciclo del segundo semestre
        if len(str(cicloinicio)) == len(str(ultciclo)):
            ciclo_segundo_semestre = cicloinicio + 9
        else:            
            ciclo_segundo_semestre = cicloinicio + 1009     # Caso especial para ciclos antes del 2000  
        
        #print(f"{ciclo_segundo_semestre}")  

        # Si el ciclo del segundo semestre corresponde a su último ciclo registrado
        # y no es igual al ciclo de egreso, 
        # entonces se considera que desertó        # 
        if (ciclo_segundo_semestre == ultciclo) and (ciclo_segundo_semestre != cicloegreso):
            return 1
        else:
            return 0
    except Exception as e:
        print(f"Error: {cicloinicio},{ciclo_segundo_semestre},{ultciclo}: {e}")
        return None
```

```python
# Aplicar la función a las filas del DataFrame
df_indicador['DESERTO'] = df_indicador.apply(
    lambda row: 
        identificar_desercion(row['CICLOINICIO'], row['ULTCICLO'], row['CICLOEGRESO']), 
    axis=1
)
```
