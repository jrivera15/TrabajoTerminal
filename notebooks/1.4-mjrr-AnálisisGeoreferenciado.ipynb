```python
# plotly stack bar chart por sexo y origen
import plotly.express as px

df_origen_sexo["FORANEO"] = df_origen_sexo["FORANEO"].astype(str)

fig = px.bar(
    df_origen_sexo, 
    x="SEXO", 
    y="TOTAL",     
    color="FORANEO",     
    color_discrete_map={'0': 'blue', '1': 'green'},
    barmode="stack", 
    text='PCT_TOTAL',
    
    )

fig.update_traces(
    texttemplate='%{text:.1f}%', 
    textposition='outside'
    )

fig.update_layout(
    title_text='Número de estudiantes inscritos por <b>sexo y residencia</b> en los últimos 5 años<br>[2202-2242]',
    height=600,
    )

fig.show()
```

### Carreras a las que se inscribieron las mujeres y los hombres

```python
df_carreras_sexo = df.groupby(['SEXO', 'CARRERA', 'NOMBRECARRERA']).size().reset_index(name='TOTAL')
df_carreras_sexo.sort_values(by='TOTAL', ascending=True, inplace=True)
df_carreras_sexo['PCT_PARCIAL'] = df_carreras_sexo['TOTAL'] / df_carreras_sexo.groupby('CARRERA')['TOTAL'].transform('sum') * 100
df_carreras_sexo['PCT_GLOBAL'] = df_carreras_sexo['TOTAL'] / df_carreras_sexo['TOTAL'].sum() * 100
df_carreras_sexo
```

```python
import plotly.express as px

# Calcular la altura dinámica en función del número de elementos en el eje Y
num_elementos_y = len(df_carreras_sexo['NOMBRECARRERA'].unique())  # Número de elementos únicos en el eje Y
altura_por_elemento = 20  # Altura en píxeles por cada elemento en el eje Y
altura_total = num_elementos_y * altura_por_elemento

# Crear gráfico
fig = px.bar(df_carreras_sexo, 
             x='TOTAL', 
             y='NOMBRECARRERA', 
             color='SEXO', 
             orientation='h',  # Apilar horizontalmente
             title="Carreras en las que se inscribieron <b>mujeres y hombres</b> en los últimos 5 años<br>[2202-2242]",
             height=altura_total,  # Altura del gráfico
             labels={'SEXO': 'Sexo','TOTAL': 'Total', },  # Etiquetas de los ejes
             text='PCT_PARCIAL',  # Mostrar el valor de cada barra
             custom_data=['SEXO'], 
)

# Personalizar el tooltip
fig.update_traces(
    texttemplate="%{text:,.1f}%",
    hovertemplate='<b>%{y}</b><br>' +
                  'Sexo: %{customdata[0]}<br>' +
                  'Total: %{x}<br>' +
                  'Pct: %{text:.1f}%' +
                  '<extra></extra>'
)

fig.show()
```

#### Carreras a las que se inscriben los alumnos locales, por código postal

```python
# Cantidad de alumnos locales inscritos por código postal y carrera
locales_carreras_cp = df_locales.groupby(['CDPOSTAL','LAT','LON','CARRERA','NOMBRECARRERA']).size().reset_index(name='cantidad')

# Muy impotante ordenar los valores aquí, para que la siguiente lista esté ordenada y poder tomar el TOP K
locales_carreras_cp.sort_values(by='cantidad', ascending=False, inplace=True)

# Crear lista de carreras por código postal y tomar el TOP 5
locales_lista_carreras_cp = locales_carreras_cp.groupby(['CDPOSTAL', 'LAT', 'LON']).agg({
    'CARRERA': lambda x: ', '.join(list(x)[:5]),  # Convertir a cadena separada por comas y tomar el TOP 5
    'NOMBRECARRERA': lambda x: ', '.join(list(x)[:5]),  # Convertir a cadena separada por comas y tomar el TOP 5
    'cantidad': 'sum',  
}).reset_index()
locales_lista_carreras_cp.sort_values(by='cantidad', ascending=False, inplace=True)
locales_lista_carreras_cp
```

```python
import plotly.express as px

fig = px.scatter_map(
    locales_lista_carreras_cp, 
    lat="LAT", 
    lon="LON",     
    color="cantidad", 
    size="cantidad",
    color_continuous_scale=px.colors.diverging.RdYlGn, 
    size_max=15, 
    zoom=9,
    title='TOP 5 carreras a las que se inscribieron los <b>alumnos locales</b> en los últimos 5 años, por código postal<br>[2202-2242]',
    hover_name='CDPOSTAL',
    hover_data=['CARRERA'],  # Columnas adicionales para el tooltip
    height=600,
    center=dict(lat=29.007553, lon=-111.333497), # Centrar el mapa
    
)

# Personalizar el tooltip con hovertemplate
fig.update_traces(
    hovertemplate="<b>CDPOSTAL: %{hovertext}</b><br>" +   
                  "Cantidad: %{marker.size}<br>" + 
                  "Carreras: %{customdata[0]}<br>" +  
                  "<extra></extra>" # Elimina el espacio adicional en el tooltip
)


fig.show()
```

```python
# Se agrupa por geolocalización (CP, LAT, LON) y se calcula el número de alumnos
df_cp = df.groupby(['CDPOSTAL', 'LAT', 'LON']).size().reset_index(name='TOTALALUMNOS')

# Foráneos y locales
df_cp['TOTALFORANEOS'] = df.groupby(['CDPOSTAL'])['FORANEO'].sum().reset_index(name='TOTALFORANEOS')['TOTALFORANEOS']
df_cp['TOTALLOCALES'] = df_cp['TOTALALUMNOS'] - df_cp['TOTALFORANEOS']

# Mujeres y hombres
df_cp['TOTALFEMENINO'] = df.groupby(['CDPOSTAL'])['SEXO'].apply(lambda x: (x == 'F').sum()).reset_index(name='TOTALFEMENINO')['TOTALFEMENINO']
df_cp['TOTALMASCULINO'] = df_cp['TOTALALUMNOS'] - df_cp['TOTALFEMENINO']

# Se obtiene el top 5 de carreras
df_cp_topcarreras = df.groupby(['CDPOSTAL', 'LAT', 'LON']).agg({
    'CARRERA': lambda x: ', '.join(list(x)[:5]),  # Convertir a cadena separada por comas y tomar el TOP 5
    'NOMBRECARRERA': lambda x: ', '.join(list(x)[:5]),  # Convertir a cadena separada por comas y tomar el TOP 5 
}) \
.reset_index() \
.rename(columns={"CARRERA": "TOPCARRERA", "NOMBRECARRERA": "TOPNOMBRECARRERA"})#, 'IDEstudiante': 'TOTALALUMNOS'})

df_cp = df_cp.merge(df_cp_topcarreras, on=['CDPOSTAL', 'LAT', 'LON'], how='left')
```
